```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
# loading the csv data to a pandas dataframe
gold = pd.read_csv('gold_price_data.csv')
```

# correlation

```python
gold_numeric = gold.select_dtypes(include=[float, int])  # Select only numeric columns
correlation =gold_numeric.corr()
```

```python
# constructing a heatmap tp understand the correlation
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={"size":8}, cmap ="plasma")
```

```python
# correlation values of GLD
print(correlation['GLD'])
```

```python
# checking the distribution of the GLD price
sns.displot(gold['GLD'], color='Red')
```

```python
x = gold.drop(['Date', 'GLD'], axis=1)
y = gold['GLD']
```

```python
x.shape
```

```python
y.shape
```

# training :

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
regressor = RandomForestRegressor(n_estimators=200)
```

```python
# training the model
regressor.fit(x_train, y_train)
```

```python
# prediction on Test Data
test_data_prediction = regressor.predict(x_test)
print(test_data_prediction)
```

```python
# R squared
score = r2_score(y_test, test_data_prediction)
print( score)
```

```python
y_test = list(y_test)
```

```python
plt.plot(y_test, color='blue', label = 'Actual Value')
plt.plot(test_data_prediction, color='green', label='Predicted Value')
plt.title('Actual vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
